```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create dataset
df = pd.DataFrame({
    "Marks": [80, 90, np.nan, 70, np.nan, 85]
})

print("Before Handling Missing Values:")
print(df)

# Show missing values (Before)
plt.figure()
plt.title("Missing Values (Before)")
plt.imshow(df.isnull().values, cmap="gray")
plt.colorbar(label="Missing (1=True, 0=False)")
plt.show()

# Fill missing values with mean
df["Marks"] = df["Marks"].fillna(df["Marks"].mean())

print("\nAfter Handling Missing Values:")
print(df)

# Show missing values (After)
plt.figure()
plt.title("Missing Values (After)")
plt.imshow(df.isnull().values, cmap="gray")
plt.colorbar(label="Missing (1=True, 0=False)")
plt.show()
```
